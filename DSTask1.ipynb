```python
#importing cleaned dataset

df1=pd.read_csv('titanic_cleaned.csv')
```

```python
#Adding new column for no.of survivors
df1['gender_count']=((df1['Survived']==1)).astype(int)
df1.head()
```

```python
#Finding the no.of males who survived
males_survived = df1[(df1['Sex'] == 'male') & (df1['Survived'] == 1)]
males_survived.head()

```

```python
numof_males_survived = males_survived.shape[0]
numof_males_survived
```

```python
#Finding the no.of females survived
females_survived = df1[(df1['Sex'] == 'female') & (df1['Survived'] == 1)]
numof_females_survived = females_survived.shape[0]
```

```python
numof_females_survived = females_survived.shape[0]
numof_females_survived
```

```python
# Plotting the graph

import matplotlib.pyplot as plt
```

```python
# Bar graph
plt.figure(figsize=(4,6))
labels = ['Males', 'Females']
values = [numof_males_survived, numof_females_survived]
plt.bar(labels,values,color=['red','blue'])
plt.ylabel('Number of survivors')
plt.xlabel('Number of male and female')

plt.show()
```

```python
#Histogram


plt.figure(figsize=(10,6))
plt.hist(df1['Age'],bins=15,edgecolor='black',color='green')
plt.title('Age Distribution on Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
